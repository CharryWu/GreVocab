```python
vocab = vocab.drop(vocab[(vocab['序号'] =='序号') 
                         | (vocab['section'] == '上海云端教育·专注 GRE') 
                         | (vocab['题号'] == '未经作者允许，禁止任何个人或机构把本词表用于商业目的')
                         | (vocab['单词'] == '未经作者允许，禁止任何个人或机构把本词表用于商业目的')].index)
```

```python
for index,row in vocab.iterrows():
    serial=row['序号']
    if type(serial) is str and not serial.isdigit():
        vocab.set_value(index,'释义',serial) #将序号那一栏的数据转移到释义拦
        vocab.set_value(index,'序号',np.NaN) # 
```

```python
vocab=vocab.dropna(how='all')
```

```python
#合并所有的单释义的行
for index,row in vocab.iterrows():
    if index == 0:
        continue
    prevrow = vocab.iloc[index-1]
    if not pd.isnull(row['释义']) and not ('3' in row['释义']) and pd.isnull(row['音标']):
        if not pd.isnull(prevrow['释义']) and ('2' in prevrow['释义']) and pd.isnull(prevrow['音标']):
            vocab.set_value(index-1,'释义',prevrow['释义']+row['释义'])
            vocab.set_value(index,'释义',np.NaN)
            print(index)
```

```python
#合并所有的多释义的行
for index,row in vocab.iterrows():
    if index == 0:
        continue
    prevrow = vocab.iloc[index-1]
    if not pd.isnull(row['释义']) and not ('3' in row['释义']) and pd.isnull(row['音标']):
        if not pd.isnull(prevrow['释义']) and ('2' in prevrow['释义']) and pd.isnull(prevrow['音标']):
            vocab.set_value(index-1,'释义',prevrow['释义']+row['释义'])
            vocab.set_value(index,'释义',np.NaN)
            print(index)
```

```python
#删除所有版权信息
mark=r"【上海云端教育·专注 GRE：添加微信 YunMengZe1020 获取机经】"
for index,row in vocab.iterrows():
    newstr=re.sub(mark,"",str(row['释义']))
    vocab.set_value(index,'释义',newstr)
```

```python
vocab.head(100)
```

```python
#将所有的在unnamed:6列的内容搬移到释义列去
for index,row in vocab.iterrows():
    if pd.isnull(row['释义']) and not pd.isnull(row['Unnamed: 6']):
        vocab.set_value(index,'释义',row['Unnamed: 6'])
        vocab.set_value(index,'Unnamed: 6',np.NaN)
```

```python
# 合并所有单释义的行
digit=re.compile('[0-9].+')
for index,row in vocab.iterrows():
    if index == 0:
        continue
    prevrow = vocab.iloc[index-1]
    if not pd.isnull(row['释义']) and pd.isnull(row['音标'])\
    and pd.isnull(row['Unnamed: 6']) and not pd.isnull(prevrow['释义'])\
    and not digit.match(row['释义']) and not digit.match(prevrow['释义']):
        vocab.set_value(index-1,'释义',prevrow['释义']+row['释义'])
        vocab.set_value(index,'释义',np.NaN)
```

```python
for index,row in vocab.iterrows():
    if pd.isnull(row['释义']):
        if not pd.isnull(row['音标']):
            vocab.set_value(index,'释义',row['音标'])
        elif not pd.isnull(row['Unnamed: 6']):
            vocab.set_value(index,'释义',row['Unnamed: 6'])
        print(index)
```
